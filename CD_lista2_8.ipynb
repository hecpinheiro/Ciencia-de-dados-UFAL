```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import matplotlib.dates as mdates
from datetime import date
```

```python
dataDrugs = pd.read_csv('https://raw.githubusercontent.com/bapimentel/Ciencia-de-Dados/master/Dados/Drugs.csv')

dataDrugs['date'] = pd.to_datetime(dataDrugs['date'])

dataDrugs
```

```python
dataDrugs.info()
```

```python
plt.figure(figsize=(20,8))
plt.plot(dataDrugs['date'], dataDrugs['value'])

plt.xticks(dataDrugs.date[5::4],size='small',rotation=35, ha="right")

plt.xlabel("Data", size = 16)
plt.ylabel("Valor", size = 16)
plt.title('Vendas de medicamentos na Austrália entre os anos de 1991 e 2008')
```
